```python
# runner.py
import subprocess
import time

RUNTIME = 4 * 60 * 60  # 4시간(초)



while True:
    print("automation.py 시작")
    # stdout과 stderr를 실시간으로 받아서 터미널에 출력
    proc = subprocess.Popen(
        ["python", "automation.py"],
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
        bufsize=1,
        text=True
    )

    try:
        start_time = time.time()
        # 실시간 출력 읽기 루프
        while True:
            # 한 줄씩 읽어서 출력
            line = proc.stdout.readline()
            if line:
                print("[automation.py]", line, end="")  # 이미 줄바꿈 포함됨
            # 프로세스가 끝나면 break
            if proc.poll() is not None:
                # 남아있는 출력 다 flush
                for line in proc.stdout:
                    print("[automation.py]", line, end="")
                break
            # 타임아웃 체크
            if time.time() - start_time > RUNTIME:
                raise subprocess.TimeoutExpired(cmd="automation.py", timeout=RUNTIME)
        # 여기까지 왔다는 것은:
        #  - RUNTIME 이내에 automation.py가 스스로 종료함
        print("automation.py가 자체 종료됨, 즉시 재시작")
    except subprocess.TimeoutExpired:
        print("4시간 경과, automation.py 강제 종료 후 재시작")
        proc.terminate()
        try:
            proc.wait(timeout=30)  # 30초 안에 정상 종료 기다림
        except subprocess.TimeoutExpired:
            proc.kill()           # 안 죽으면 강제 종료
        # 남아있는 출력 다 flush
        if proc.stdout:
            for line in proc.stdout:
                print("[automation.py]", line, end="")
        # 이후 while True 루프 다음 반복에서 재실행
```
